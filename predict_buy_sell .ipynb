```python
def abc(c):
    l=c.count('Buy')
    m=c.count('Sell')
    if l > 4 :
        tt='Strong Buy'
    elif l > m and l <= 4  and l > 1:
        tt='Buy'
    elif m > 4:
        tt='Strong Sell'
    elif m > l and m <= 4 and m > 1 :
        tt='Sell'
    else:
        tt='Neutral'
    return tt
```
