```python
# Commodity Channel Index 
def CCI(data, ndays):
    c=0
    cci=[]
    prices=[]
    TP = (data['high'] + data['low'] + data['close']) / 3 
    CCI = pd.Series((TP - TP.rolling(ndays).mean()) / (0.015 * TP.rolling(ndays).std()),
                    name = 'CCI') 
    data = data.join(CCI)
    
    while c < len(data):    
        prices.append(data.iloc[c,4])
        cci.append(data.iloc[c,6])
        c+=1
    return prices,cci,data
```
